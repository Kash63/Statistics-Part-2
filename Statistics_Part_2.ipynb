3. What is the significance level in hypothesis testing, and why is it important?

> Why is the Significance Level Important?
Controls the Risk of Type I Error: This is the most direct and fundamental reason. By setting α, you are explicitly defining the maximum acceptable probability of making a "false positive" conclusion. Establishes a Decision Rule Before Data Analysis: It's crucial to set the significance level before you conduct your experiment or analyze your data.  Balances Type I and Type II Errors: There's an inherent trade-off between Type I errors (α) and Type II errors (β, the probability of failing to reject a false null hypothesis).

The choice of α depends on the specific context and the relative costs of making each type of error. Enhances Comparability and Reproducibility: Using a consistent significance level across studies in a particular field allows for easier comparison of results and contributes to the reproducibility of scientific findings.

4. What does a P-value represent in hypothesis testing?
> n hypothesis testing, the p-value (or probability value) is a measure of the evidence against the null hypothesis. It quantifies how likely it is to observe a sample result as extreme as, or more extreme than, the one you obtained, assuming that the null hypothesis is true.

5. How do you interpret the P-value in hypothesis testing?
> In hypothesis testing, the p-value represents the probability of observing a test result as extreme as, or more extreme than, the one obtained from your sample data, assuming that the null hypothesis is true. A small p-value (typically ≤ the chosen significance level, α) suggests that your observed data would be very unlikely to occur if the null hypothesis were correct, thus providing strong evidence to reject the null hypothesis in favor of the alternative hypothesis. Conversely, a large p-value (> α) indicates that your observed data is reasonably likely to occur even if the null hypothesis were true, meaning there isn't sufficient evidence to reject the null hypothesis, and you fail to conclude a statistically significant effect or difference.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing/
> The choice between a one-tailed and a two-tailed test in hypothesis testing hinges on the directionality of the alternative hypothesis. A one-tailed (or directional) test is used when the alternative hypothesis predicts a specific direction of an effect or difference (e.g., the new drug will increase recovery rates, or the average score is less than a certain value). In this case, the rejection region for the null hypothesis is entirely in one tail of the sampling distribution (either the upper or lower tail). Conversely, a two-tailed (or non-directional) test is used when the alternative hypothesis simply states that there is a difference or effect, without specifying its direction (e.g., the new drug will change recovery rates, or the average score is different from a certain value). Here, the rejection region is split equally between both tails of the sampling distribution, accounting for deviations in either the positive or negative direction.

8.  What is the Z-test, and when is it used in hypothesis testing?
> The Z-test is a statistical hypothesis test used to determine if there's a significant difference between a sample mean and a population mean, or between the means of two populations. It is primarily used when the population standard deviation is known, and typically when the sample size is large (generally n≥30). The Z-test relies on the assumption that the data are normally distributed or that the sample size is large enough for the Central Limit Theorem to apply, which ensures the sampling distribution of the sample mean is approximately normal.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
> The Z-score, also known as a standard score, is a measure of how many standard deviations an observed value (like a sample mean) is away from the hypothesized population mean. It's calculated using the formula: Z=(X−μ)/(σ/
n), where X is the sample mean, μ is the hypothesized population mean, σ is the population standard deviation, and n is the sample size. In hypothesis testing, the Z-score transforms your sample data into a standardized scale, allowing you to compare it to a standard normal distribution. A large absolute Z-score indicates that your sample mean is far from the hypothesized population mean, providing evidence against the null hypothesis, while a Z-score close to zero suggests the sample mean is consistent with the null hypothesis.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
> The T-distribution, also known as Student's t-distribution, is a continuous probability distribution that resembles the bell-shaped normal distribution but has heavier tails. This means it gives more probability to extreme values, reflecting greater uncertainty. It should be used instead of the normal distribution (Z-distribution) primarily when the population standard deviation is unknown and must be estimated from the sample data. Additionally, it is generally preferred for smaller sample sizes (typically n<30), as the uncertainty in estimating the population standard deviation is more pronounced with fewer observations. As the sample size (and thus the degrees of freedom) increases, the T-distribution approaches the shape of the standard normal distribution, making the distinction less critical for very large samples.

11. What is the difference between a Z-test and a T-test?
> The fundamental difference between a Z-test and a T-test lies in their assumptions about the population standard deviation and the sample size. A Z-test is typically employed when the population standard deviation is known, or when the sample size is large (generally n≥30), allowing the use of the known population standard deviation or relying on the Central Limit Theorem to approximate a normal distribution. In contrast, a T-test is used when the population standard deviation is unknown and must be estimated from the sample data, and is particularly suited for smaller sample sizes (typically n<30). The T-distribution, which the T-test utilizes, accounts for the increased uncertainty that arises from estimating the population standard deviation from a limited number of observations, featuring heavier tails than the normal distribution.

12. What is the T-test, and how is it used in hypothesis testing?
> The T-test is a widely used statistical hypothesis test that compares the means of two groups or a sample mean to a known population mean. It's particularly valuable in hypothesis testing when the population standard deviation is unknown and must be estimated from the sample data, and it's generally preferred for smaller sample sizes. By calculating a t-statistic, which measures the difference between sample means relative to the variability within the samples, and then comparing it to a critical value from the t-distribution (based on degrees of freedom), the T-test helps determine if any observed differences are statistically significant or merely due to random chance, allowing researchers to draw conclusions about the population.

13.  What is the relationship between Z-test and T-test in hypothesis testing?
> The Z-test and T-test are both parametric hypothesis tests used to compare means, but their relationship is largely defined by the information available about the population and the sample size. Essentially, the T-test is a more robust and frequently used alternative to the Z-test, especially when dealing with real-world scenarios where population parameters are unknown. When the population standard deviation is known and/or the sample size is large (n≥30), the Z-test is applicable because the Central Limit Theorem ensures the sample mean distribution approximates a normal distribution. However, when the population standard deviation is unknown and must be estimated from the sample (using the sample standard deviation), particularly with smaller sample sizes (n<30), the T-test is necessary. The T-distribution, with its fatter tails, accounts for the increased uncertainty that comes from estimating the population standard deviation, effectively converging to the Z-distribution as the sample size grows.

14. What is a confidence interval, and how is it used to interpret statistical results?
> A confidence interval is a range of values, calculated from sample data, that is likely to contain the true value of an unknown population parameter (like the population mean or proportion) with a certain degree of confidence. For example, a 95% confidence interval for a mean means that if you were to repeat your sampling process many times, 95% of the confidence intervals you construct would contain the true population mean. It is used to interpret statistical results by providing an estimate of the precision and reliability of your sample estimate: a narrower interval suggests a more precise estimate, while a wider interval indicates more uncertainty. Crucially, if a confidence interval for a difference or effect (e.g., the difference between two means) includes zero, it implies that there is no statistically significant difference at the chosen confidence level, aligning with the idea of failing to reject the null hypothesis.

15. What is the margin of error, and how does it affect the confidence interval?
> The margin of error (MOE) is a critical component of a confidence interval, representing the range around a sample estimate within which the true population parameter is likely to fall. It quantifies the inherent uncertainty due to using a sample instead of the entire population. The confidence interval is constructed by taking the sample estimate and adding/subtracting the margin of error (e.g., Sample Mean ± Margin of Error). A larger margin of error results in a wider confidence interval, indicating less precision in the estimate and more uncertainty about the true population parameter. Conversely, a smaller margin of error leads to a narrower confidence interval, signifying a more precise estimate and greater confidence in the proximity of the sample estimate to the true population value. The margin of error is primarily affected by the sample size (larger sample = smaller MOE), the variability in the population (higher variability = larger MOE), and the chosen confidence level (higher confidence level = larger MOE).

16. How is Bayes' Theorem used in statistics, and what is its significance?
> Bayes' Theorem is a fundamental principle in statistics that provides a mathematical framework for updating the probability of an event or hypothesis based on new evidence or information. It allows us to combine our initial beliefs (prior probabilities) with observed data (likelihood) to produce a revised, more informed probability (posterior probability). Its significance lies in forming the bedrock of Bayesian statistics, a powerful approach that stands in contrast to frequentist methods by directly quantifying the probability of hypotheses. This enables a more intuitive interpretation of results, better decision-making under uncertainty, and wide-ranging applications in fields like medical diagnosis, spam filtering, machine learning, and artificial intelligence, where evidence continuously refines our understanding of the world.

17. What is the Chi-square distribution, and when is it used?
> The Chi-square (χ
2
 ) distribution is a continuous probability distribution that is non-negative and typically skewed to the right, with its shape determined by its "degrees of freedom." It is primarily used in hypothesis testing for categorical data, particularly to assess relationships between observed and expected frequencies. 18. What is the Chi-square goodness of fit test, and how is it applied?
> The Chi-square goodness-of-fit test is a statistical hypothesis test used to determine whether an observed frequency distribution of a single categorical variable significantly differs from a hypothesized or expected frequency distribution. It is applied by comparing the observed counts in each category of your sample data to the expected counts that would occur if your hypothesized distribution were true. The test calculates a chi-square statistic, which measures the discrepancy between observed and expected frequencies. If this calculated statistic is large enough, suggesting a substantial deviation from the expected distribution, and its corresponding p-value falls below the chosen significance level, then the null hypothesis (that the observed distribution fits the expected distribution) is rejected.

19. What is the F-distribution, and when is it used in hypothesis testing?
> The F-distribution is a continuous probability distribution that is non-negative and typically right-skewed, characterized by two distinct degrees of freedom: one for the numerator and one for the denominator. It is primarily used in hypothesis testing to compare variances or to assess the ratio of variances between two or more populations. The most common application is in Analysis of Variance (ANOVA), where it's used to determine if there are statistically significant differences between the means of three or more groups by comparing the variance between group means to the variance within groups. It's also used in regression analysis to test the overall significance of a regression model or to compare the fit of nested models.

20. What is an ANOVA test, and what are its assumptions?
> An ANOVA (Analysis of Variance) test is a statistical method used to determine if there are statistically significant differences between the means of three or more independent groups. Instead of performing multiple individual t-tests, which would inflate the Type I error rate, ANOVA analyzes the variance within and between groups to calculate an F-statistic. This F-statistic helps determine if the observed differences in group means are likely due to a true effect or simply random chance. The primary assumptions for an ANOVA test are: 1) Independence of observations, meaning data points within and between groups are unrelated; 2) Normality, where the data within each group should be approximately normally distributed; and 3) Homogeneity of variances, meaning the variability (spread) of data within each group should be roughly equal.

21. What are the different types of ANOVA tests?
> ANOVA offers various types to accommodate different experimental designs and research questions. The most common types include One-Way ANOVA, used when comparing the means of three or more groups based on a single categorical independent variable; Two-Way ANOVA, which examines the effects of two independent categorical variables on a dependent variable, including their potential interaction; Factorial ANOVA, an extension of two-way ANOVA for designs with more than two independent categorical variables; Repeated Measures ANOVA, employed when the same subjects are measured multiple times under different conditions or over time; ANCOVA (Analysis of Covariance), which extends ANOVA by including one or more continuous covariates to control for their influence on the dependent variable; and MANOVA (Multivariate Analysis of Variance), used when comparing the means of two or more groups on two or more dependent variables simultaneously. Each type addresses specific research scenarios by partitioning variance in different ways.

22. What is the F-test, and how does it relate to hypothesis testing?
> The F-test is a statistical hypothesis test that fundamentally examines the ratio of two variances. It is a cornerstone of various statistical analyses, most notably Analysis of Variance (ANOVA) and certain types of regression analysis. In hypothesis testing, the F-test allows us to determine if there is a statistically significant difference between the variances of two populations (e.g., to check for homogeneity of variances before conducting a t-test) or, more commonly, to assess if the means of three or more groups are significantly different. In ANOVA, the F-statistic is calculated as the ratio of the "between-group variance" (variation among group means) to the "within-group variance" (variation within each group). A large F-statistic, along with a small p-value, suggests that the variation between group means is significantly larger than what would be expected by random chance, leading to the rejection of the null hypothesis that all group means are equal.

```python
'''1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.'''
import numpy as np
from statsmodels.stats.weightstats import ztest as ztest_statmodels

population_mean_height = 170
population_std_dev = 5

sample_heights = np.array([171, 175, 168, 173, 170, 172, 176, 169, 174, 171,
                           170, 172, 175, 167, 173, 170, 171, 174, 172, 169,
                           173, 170, 175, 171, 172])

n = len(sample_heights)
sample_mean_height = np.mean(sample_heights)

alpha = 0.05
z_statistic, p_value = ztest_statmodels(x1=sample_heights, value=population_mean_height,alternative='two-sided', ddof=0)

print(f"Population Mean Height (μ): {population_mean_height} cm")
print(f"Population Standard Deviation (σ): {population_std_dev} cm")
print(f"Sample Size (n): {n}")
print(f"Sample Mean Height: {sample_mean_height:.2f} cm")
print(f"Calculated Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (α): {alpha}\n")

print("--- Hypothesis Test Conclusion ---")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}),")
    print("we reject the Null Hypothesis (H0).")
    print("This suggests that the sample mean height is significantly different from the population mean height.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}),")
    print("we fail to reject the Null Hypothesis (H0).")
    print("This suggests that there is not enough evidence to conclude that the sample mean height is significantly different from the population mean height.")
```

```python
'''2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python. '''
import numpy as np
from scipy import stats

np.random.seed(42)

sample_size_1 = 30
sample_size_2 = 35
population_mean_1 = 100
population_mean_2 = 103
population_std_dev = 10

data_group1 = np.random.normal(loc=population_mean_1, scale=population_std_dev, size=sample_size_1)
data_group2 = np.random.normal(loc=population_mean_2, scale=population_std_dev, size=sample_size_2)

t_statistic, p_value = stats.ttest_ind(data_group1, data_group2, equal_var=True)

alpha = 0.05

print(f"Mean of Group 1: {np.mean(data_group1):.2f}")
print(f"Mean of Group 2: {np.mean(data_group2):.2f}")
print(f"Calculated T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (α): {alpha}\n")

print("--- Hypothesis Test Conclusion ---")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}),")
    print("we reject the Null Hypothesis.")
    print("This suggests a statistically significant difference between the means of the two groups.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}),")
    print("we fail to reject the Null Hypothesis.")
    print("This suggests there is no statistically significant difference between the means of the two groups.")
```

```python
''' 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot. '''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from statsmodels.stats.weightstats import ztest

np.random.seed(42)

population_mean = 100
population_std_dev = 10

sample_size = 50
sample_data = np.random.normal(loc=102.5, scale=population_std_dev, size=sample_size)

sample_mean = np.mean(sample_data)

alpha = 0.05

z_statistic, p_value = ztest(x1=sample_data, value=population_mean)

z_critical_lower = norm.ppf(alpha / 2)
z_critical_upper = norm.ppf(1 - alpha / 2)

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', lw=2)

x_reject_lower = x[x < z_critical_lower]
y_reject_lower = norm.pdf(x_reject_lower, 0, 1)
plt.fill_between(x_reject_lower, 0, y_reject_lower, color='red', alpha=0.3, label=f'Rejection Region (Lower {alpha/2*100:.1f}%)')

x_reject_upper = x[x > z_critical_upper]
y_reject_upper = norm.pdf(x_reject_upper, 0, 1)
plt.fill_between(x_reject_upper, 0, y_reject_upper, color='red', alpha=0.3, label=f'Rejection Region (Upper {alpha/2*100:.1f}%)')

plt.axvline(z_statistic, color='green', linestyle='--', lw=2, label=f'Calculated Z-statistic: {z_statistic:.2f}')
plt.axvline(z_critical_lower, color='black', linestyle=':', lw=1, label=f'Critical Z ({alpha/2:.2f}): {z_critical_lower:.2f}')
plt.axvline(z_critical_upper, color='black', linestyle=':', lw=1, label=f'Critical Z ({1-alpha/2:.2f}): {z_critical_upper:.2f}')


plt.title('Two-Tailed Z-Test: Decision Region')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()

print(f"Population Mean: {population_mean}")
print(f"Population Standard Deviation: {population_std_dev}")
print(f"Sample Size: {sample_size}")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Calculated Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (α): {alpha}\n")

print("--- Hypothesis Test Conclusion ---")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}),")
    print("we reject the Null Hypothesis (H0).")
    print("This suggests that the sample mean is significantly different from the population mean.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}),")
    print("we fail to reject the Null Hypothesis (H0).")
    print("This suggests that there is no significant evidence that the sample mean is different from the population mean.")
```

```python

'''5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_and_visualize_errors(
    population_mean_null,
    population_std_dev,
    sample_size,
    alpha,
    true_population_mean_alt
):

    std_error = population_std_dev / np.sqrt(sample_size)


    z_critical_lower = norm.ppf(alpha / 2)
    z_critical_upper = norm.ppf(1 - alpha / 2)


    x = np.linspace(population_mean_null - 4 * population_std_dev, population_mean_null + 4 * population_std_dev, 1000)

    y_null = norm.pdf(x, population_mean_null, std_error)


    x_type1_lower = x[x < population_mean_null + z_critical_lower * std_error]
    y_type1_lower = norm.pdf(x_type1_lower, population_mean_null, std_error)

    x_type1_upper = x[x > population_mean_null + z_critical_upper * std_error]
    y_type1_upper = norm.pdf(x_type1_upper, population_mean_null, std_error)


    y_alt = norm.pdf(x, true_population_mean_alt, std_error)


    x_type2 = x[x >= population_mean_null + z_critical_lower * std_error]
    x_type2 = x_type2[x_type2 <= population_mean_null + z_critical_upper * std_error]
    y_type2 = norm.pdf(x_type2, true_population_mean_alt, std_error)


    beta_error = norm.cdf(population_mean_null + z_critical_upper * std_error, loc=true_population_mean_alt, scale=std_error) - \
                 norm.cdf(population_mean_null + z_critical_lower * std_error, loc=true_population_mean_alt, scale=std_error)

    power = 1 - beta_error


    plt.figure(figsize=(12, 7))
    plt.plot(x, y_null, color='blue', lw=2, label='Null Hypothesis Distribution (H0)')
    plt.plot(x, y_alt, color='orange', lw=2, label='Alternative Hypothesis Distribution (H1)')


    plt.fill_between(x_type1_lower, 0, y_type1_lower, color='red', alpha=0.3, label=f'Type I Error (α/2): {alpha/2:.4f}')
    plt.fill_between(x_type1_upper, 0, y_type1_upper, color='red', alpha=0.3)

    plt.fill_between(x_type2, 0, y_type2, color='green', alpha=0.3, label=f'Type II Error (β): {beta_error:.4f}')

    plt.title('Type I and Type II Errors Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.axvline(population_mean_null + z_critical_lower * std_error, color='grey', linestyle='--', label='Critical Values')
    plt.axvline(population_mean_null + z_critical_upper * std_error, color='grey', linestyle='--')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
```

```python
'''9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.'''
import numpy as np
from scipy import stats

def calculate_confidence_interval(sample_data, confidence_level=0.95):
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    standard_error = sample_std / np.sqrt(sample_size)

    degrees_freedom = sample_size - 1

    alpha = 1 - confidence_level

    t_critical = stats.t.ppf(1 - alpha / 2, degrees_freedom)

    margin_of_error = t_critical * standard_error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

np.random.seed(42)
sample_data = np.random.normal(loc=75, scale=10, size=50)

confidence_level = 0.95
lower, upper = calculate_confidence_interval(sample_data, confidence_level)

print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"Sample Standard Deviation: {np.std(sample_data, ddof=1):.2f}")
print(f"Sample Size: {len(sample_data)}")
print(f"{int(confidence_level*100)}% Confidence Interval for the Sample Mean: ({lower:.2f}, {upper:.2f})")
```

```python
''' 11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process. '''
def bayesian_inference(prior_h, likelihood_e_given_h, likelihood_e_given_not_h):
    prior_not_h = 1 - prior_h

    marginal_e = (likelihood_e_given_h * prior_h) + (likelihood_e_given_not_h * prior_not_h)

    posterior_h_given_e = (likelihood_e_given_h * prior_h) / marginal_e

    return posterior_h_given_e, marginal_e

prevalence_disease = 0.01
test_sensitivity = 0.95
test_false_positive_rate = 0.10

posterior_probability, probability_evidence = bayesian_inference(
    prevalence_disease, test_sensitivity, test_false_positive_rate
)

print(f"Prior Probability of Disease (P(H)): {prevalence_disease:.4f}")
print(f"Likelihood of Positive Test given Disease (P(E|H)): {test_sensitivity:.4f}")
print(f"Likelihood of Positive Test given No Disease (P(E|~H)): {test_false_positive_rate:.4f}")
print(f"Probability of Positive Test (P(E)): {probability_evidence:.4f}\n")
print(f"Posterior Probability of Disease given Positive Test (P(H|E)): {posterior_probability:.4f}")
```

```python
'18. Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results.'
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import f_oneway

np.random.seed(42)

group_a = np.random.normal(loc=70, scale=10, size=30)
group_b = np.random.normal(loc=75, scale=10, size=30)
group_c = np.random.normal(loc=70, scale=10, size=30)
group_d = np.random.normal(loc=85, scale=10, size=30)

f_statistic, p_value = f_oneway(group_a, group_b, group_c, group_d)

alpha = 0.05

print(f"Mean of Group A: {np.mean(group_a):.2f}")
print(f"Mean of Group B: {np.mean(group_b):.2f}")
print(f"Mean of Group C: {np.mean(group_c):.2f}")
print(f"Mean of Group D: {np.mean(group_d):.2f}\n")
print(f"Calculated F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (α): {alpha}\n")

print("--- Hypothesis Test Conclusion ---")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}),")
    print("we reject the Null Hypothesis.")
    print("This suggests that there is a statistically significant difference between the means of at least two of the groups.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}),")
    print("we fail to reject the Null Hypothesis.")
    print("This suggests that there is no statistically significant difference between the means of the groups.")

data_combined = {
    'Score': np.concatenate([group_a, group_b, group_c, group_d]),
    'Group': ['A'] * len(group_a) + ['B'] * len(group_b) + ['C'] * len(group_c) + ['D'] * len(group_d)
}
df = pd.DataFrame(data_combined)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Score', data=df, palette='viridis')
sns.swarmplot(x='Group', y='Score', data=df, color='black', alpha=0.7)
plt.title('Distribution of Scores Across Different Groups')
plt.xlabel('Group')
plt.ylabel('Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
'19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.'
import numpy as np
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups, alpha=0.05):
    print("--- Checking ANOVA Assumptions ---")

    print("\n1. Normality Assumption (Shapiro-Wilk Test for each group):")
    normality_results = {}
    for i, group_data in enumerate(groups):
        stat, p = shapiro(group_data)
        normality_results[f'Group {i+1}'] = {'Statistic': stat, 'P-value': p}
        if p < alpha:
            print(f"  Group {i+1}: P-value = {p:.4f} < {alpha} (Reject H0 - Not normal)")
        else:
            print(f"  Group {i+1}: P-value = {p:.4f} >= {alpha} (Fail to reject H0 - Appears normal)")
    print("  Note: For large sample sizes, ANOVA is robust to minor normality violations.")

    print("\n2. Independence Assumption:")
    print("  This assumption is assessed based on the experimental design.")
    print("  Observations within each group and between groups should be independent (e.g., different subjects in each group).")
    print("  No statistical test directly confirms independence from the data alone.")

    print("\n3. Homogeneity of Variances Assumption (Levene's Test):")
    stat, p = levene(*groups)
    print(f"  Levene's Test Statistic: {stat:.4f}, P-value: {p:.4f}")
    if p < alpha:
        print(f"  P-value = {p:.4f} < {alpha} (Reject H0 - Variances are NOT equal)")
    else:
        print(f"  P-value = {p:.4f} >= {alpha} (Fail to reject H0 - Variances appear equal)")
    print("  Note: If variances are not equal, consider Welch's ANOVA or data transformation.")

np.random.seed(42)
group1 = np.random.normal(loc=10, scale=2, size=30)
group2 = np.random.normal(loc=12, scale=2.1, size=30)
group3 = np.random.normal(loc=10, scale=2.5, size=30)

check_anova_assumptions(group1, group2, group3)
```

```python
'''20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
results.'''
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

n_per_group = 25

data = {
    'FactorA': ['X'] * n_per_group + ['X'] * n_per_group + ['Y'] * n_per_group + ['Y'] * n_per_group,
    'FactorB': ['P'] * n_per_group + ['Q'] * n_per_group + ['P'] * n_per_group + ['Q'] * n_per_group,
    'Score': np.concatenate([
        np.random.normal(loc=60, scale=8, size=n_per_group),
        np.random.normal(loc=70, scale=8, size=n_per_group),
        np.random.normal(loc=65, scale=8, size=n_per_group),
        np.random.normal(loc=55, scale=8, size=n_per_group)
    ])
}
df = pd.DataFrame(data)

model = ols('Score ~ C(FactorA) * C(FactorB)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

alpha = 0.05

print("--- Two-Way ANOVA Results ---")
print(anova_table)
print(f"\nSignificance Level (α): {alpha}\n")

print("--- Hypothesis Test Conclusion ---")

p_value_factor_a = anova_table.loc['C(FactorA)', 'PR(>F)']
if p_value_factor_a < alpha:
    print(f"FactorA Main Effect (p={p_value_factor_a:.4f}): Statistically significant difference across levels of FactorA.")
else:
    print(f"FactorA Main Effect (p={p_value_factor_a:.4f}): No statistically significant difference across levels of FactorA.")

p_value_factor_b = anova_table.loc['C(FactorB)', 'PR(>F)']
if p_value_factor_b < alpha:
    print(f"FactorB Main Effect (p={p_value_factor_b:.4f}): Statistically significant difference across levels of FactorB.")
else:
    print(f"FactorB Main Effect (p={p_value_factor_b:.4f}): No statistically significant difference across levels of FactorB.")

p_value_interaction = anova_table.loc['C(FactorA):C(FactorB)', 'PR(>F)']
if p_value_interaction < alpha:
    print(f"Interaction Effect (p={p_value_interaction:.4f}): Statistically significant interaction between FactorA and FactorB.")
    print("The effect of one factor on Score depends on the level of the other factor.")
else:
    print(f"Interaction Effect (p={p_value_interaction:.4f}): No statistically significant interaction between FactorA and FactorB.")
    print("The effects of FactorA and FactorB on Score are independent of each other.")

plt.figure(figsize=(10, 6))
sns.pointplot(x='FactorA', y='Score', hue='FactorB', data=df, ci=95, palette='deep', markers='o')
plt.title('Interaction Plot of FactorA and FactorB on Score')
plt.xlabel('Factor A')
plt.ylabel('Mean Score')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Factor B')
plt.tight_layout()
plt.show()
```

```python
'''22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.'''
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import f_oneway

np.random.seed(42)

group1_data = np.random.normal(loc=10, scale=2, size=30)
group2_data = np.random.normal(loc=12, scale=2, size=30)
group3_data = np.random.normal(loc=10, scale=2, size=30)
group4_data = np.random.normal(loc=15, scale=2, size=30)

f_statistic, p_value = f_oneway(group1_data, group2_data, group3_data, group4_data)

alpha = 0.05

print(f"Mean of Group 1: {np.mean(group1_data):.2f}")
print(f"Mean of Group 2: {np.mean(group2_data):.2f}")
print(f"Mean of Group 3: {np.mean(group3_data):.2f}")
print(f"Mean of Group 4: {np.mean(group4_data):.2f}\n")
print(f"Calculated F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (α): {alpha}\n")

print("--- Hypothesis Test Conclusion ---")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}),")
    print("we reject the Null Hypothesis.")
    print("This suggests that there is a statistically significant difference between the means of at least two of the groups.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}),")
    print("we fail to reject the Null Hypothesis.")
    print("This suggests that there is no statistically significant difference between the means of the groups.")

data_combined = {
    'Value': np.concatenate([group1_data, group2_data, group3_data, group4_data]),
    'Group': ['Group 1'] * len(group1_data) + ['Group 2'] * len(group2_data) +
             ['Group 3'] * len(group3_data) + ['Group 4'] * len(group4_data)
}
df = pd.DataFrame(data_combined)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Value', data=df, palette='viridis')
sns.swarmplot(x='Group', y='Value', data=df, color='black', alpha=0.7)
plt.title('Comparison of Group Means (One-Way ANOVA)')
plt.xlabel('Group')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
'''23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.'''
import numpy as np
from scipy import stats

np.random.seed(42)

sample_size = 50
hypothesized_population_mean = 100
true_population_mean = 102
population_std_dev = 10

sample_data = np.random.normal(loc=true_population_mean, scale=population_std_dev, size=sample_size)

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

t_statistic, p_value = stats.ttest_1samp(sample_data, popmean=hypothesized_population_mean)

alpha = 0.05

print(f"Sample Size: {sample_size}")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")
print(f"Hypothesized Population Mean (H0): {hypothesized_population_mean}")
print(f"Calculated T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (α): {alpha}\n")

print("--- Hypothesis Test Conclusion ---")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}),")
    print("we reject the Null Hypothesis.")
    print(f"This suggests the sample mean is significantly different from {hypothesized_population_mean}.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}),")
    print("we fail to reject the Null Hypothesis.")
    print(f"This suggests no significant evidence that the sample mean is different from {hypothesized_population_mean}.")
```

```python
'''24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results. '''
import numpy as np
from scipy.stats import chi2

np.random.seed(42)

sample_size = 30
hypothesized_pop_variance = 25.0
true_pop_std_dev = 5.5

sample_data = np.random.normal(loc=100, scale=true_pop_std_dev, size=sample_size)

sample_variance = np.var(sample_data, ddof=1)
degrees_of_freedom = sample_size - 1

chi2_statistic = (degrees_of_freedom * sample_variance) / hypothesized_pop_variance

alpha = 0.05

p_value = chi2.cdf(chi2_statistic, degrees_of_freedom)
p_value = min(p_value, 1 - p_value) * 2

print(f"Sample Size: {sample_size}")
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Hypothesized Population Variance (H0): {hypothesized_pop_variance}")
print(f"Degrees of Freedom: {degrees_of_freedom}")
print(f"Calculated Chi-square Statistic: {chi2_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (α): {alpha}\n")

print("--- Hypothesis Test Conclusion ---")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}),")
    print("we reject the Null Hypothesis.")
    print("This suggests that the population variance is significantly different from the hypothesized value.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}),")
    print("we fail to reject the Null Hypothesis.")
    print("This suggests there is no significant evidence that the population variance is different from the hypothesized value.")
```

```python
'''25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.'''
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

successes1 = 45
nobs1 = 100

successes2 = 30
nobs2 = 100

counts = np.array([successes1, successes2])
nobs = np.array([nobs1, nobs2])

z_statistic, p_value = proportions_ztest(counts, nobs)

alpha = 0.05

print(f"Group 1 Successes: {successes1}, Total Observations: {nobs1}")
print(f"Group 2 Successes: {successes2}, Total Observations: {nobs2}")
print(f"Proportion Group 1: {successes1/nobs1:.4f}")
print(f"Proportion Group 2: {successes2/nobs2:.4f}")
print(f"Calculated Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (α): {alpha}\n")

print("--- Hypothesis Test Conclusion ---")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}),")
    print("we reject the Null Hypothesis.")
    print("This suggests a statistically significant difference between the proportions of the two groups.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}),")
    print("we fail to reject the Null Hypothesis.")
    print("This suggests there is no statistically significant difference between the proportions of the two groups.")
```

```python
'''26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

np.random.seed(42)

sample_size_1 = 40
sample_size_2 = 50
mean_1 = 100
mean_2 = 102
std_dev_1 = 8  # Population standard deviation for group 1
std_dev_2 = 12 # Population standard deviation for group 2 (larger variance)

data_group1 = np.random.normal(loc=mean_1, scale=std_dev_1, size=sample_size_1)
data_group2 = np.random.normal(loc=mean_2, scale=std_dev_2, size=sample_size_2)

variance_group1 = np.var(data_group1, ddof=1)
variance_group2 = np.var(data_group2, ddof=1)

if variance_group1 >= variance_group2:
    f_statistic = variance_group1 / variance_group2
    dfn = sample_size_1 - 1
    dfd = sample_size_2 - 1
else:
    f_statistic = variance_group2 / variance_group1
    dfn = sample_size_2 - 1
    dfd = sample_size_1 - 1

p_value = 2 * f.sf(f_statistic, dfn, dfd)

alpha = 0.05

critical_f_value_upper = f.ppf(1 - alpha / 2, dfn, dfd)

print(f"Sample 1 Size: {sample_size_1}")
print(f"Sample 1 Mean: {np.mean(data_group1):.2f}")
print(f"Sample 1 Variance: {variance_group1:.2f}\n")
print(f"Sample 2 Size: {sample_size_2}")
print(f"Sample 2 Mean: {np.mean(data_group2):.2f}")
print(f"Sample 2 Variance: {variance_group2:.2f}\n")

print(f"Calculated F-statistic: {f_statistic:.4f}")
print(f"Numerator Degrees of Freedom (df1): {dfn}")
print(f"Denominator Degrees of Freedom (df2): {dfd}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (α): {alpha}")
print(f"Upper Critical F-value (for α/2): {critical_f_value_upper:.4f}\n")


print("--- Hypothesis Test Conclusion ---")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}),")
    print("we reject the Null Hypothesis (H0: variances are equal).")
    print("This suggests a statistically significant difference between the variances of the two datasets.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}),")
    print("we fail to reject the Null Hypothesis (H0: variances are equal).")
    print("This suggests there is no statistically significant difference between the variances of the two datasets.")


x = np.linspace(0.01, max(f_statistic + 1, critical_f_value_upper + 1, 4), 500)
pdf = f.pdf(x, dfn, dfd)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, color='blue', lw=2)

x_reject_upper = x[x > critical_f_value_upper]
y_reject_upper = f.pdf(x_reject_upper, dfn, dfd)
plt.fill_between(x_reject_upper, 0, y_reject_upper, color='red', alpha=0.3, label=f'Rejection Region (Upper {alpha/2*100:.1f}%)')

plt.axvline(f_statistic, color='green', linestyle='--', lw=2, label=f'Calculated F-statistic: {f_statistic:.2f}')
plt.axvline(critical_f_value_upper, color='black', linestyle=':', lw=1, label=f'Critical F ({1-alpha/2:.2f}): {critical_f_value_upper:.2f}')

plt.title('F-Test for Comparing Variances: Decision Region')
plt.xlabel(f'F-value (df1={dfn}, df2={dfd})')
plt.ylabel('Probability Density')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
'''27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.'''
import numpy as np
from scipy.stats import chisquare

np.random.seed(42)

total_observations = 200
categories = 4

observed_frequencies = np.array([55, 45, 40, 60])

expected_frequencies = np.array([total_observations / categories] * categories)

chi2_statistic, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

alpha = 0.05

print(f"Observed Frequencies: {observed_frequencies}")
print(f"Expected Frequencies (Uniform): {expected_frequencies}")
print(f"Chi-square Statistic: {chi2_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (α): {alpha}\n")

print("--- Hypothesis Test Conclusion ---")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}),")
    print("we reject the Null Hypothesis.")
    print("This suggests that the observed data distribution is significantly different from the expected distribution.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}),")
    print("we fail to reject the Null Hypothesis.")
    print("This suggests that there is no significant difference between the observed data distribution and the expected distribution.")
```
